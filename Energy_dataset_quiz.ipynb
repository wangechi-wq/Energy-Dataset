```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading dataset
energy = pd.read_csv('/content/energydata_complete.csv')
energy.head(4)
```

```python
#checking the shape of the dataset
print(f'The dataset contains: {energy.shape[0]} rows and {energy.shape[1]} columns')
```

```python
#converting the date column to datatime data type
energy.date=pd.to_datetime(energy.date, infer_datetime_format=True)
```

```python
#checking the descriptive statistics for the dataset
energy.describe()
```

```python
#checking correlation between the variables
energy.corr()
```

we can see the lights variable has a very low correlation with the other variables so we can drop the column

```python
#we will drop the date and lights column since they are not necessary for our analysis
energy_df=energy.drop(['lights','date'], axis=1)
energy_df.head(3)
```

```python
#normalize the dataset
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaling = pd.DataFrame(scaler.fit_transform(energy_df),  columns = energy_df.columns)

#get features and labels
X = scaling.drop(columns=['Appliances'])

y = scaling['Appliances']
```

```python
#split the train and test data

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print('X_train shape: {}'.format(X_train.shape))
print('y_train shape: {}'.format(y_train.shape))
print('X_test shape: {}'.format(X_test.shape))
print('y_test shape: {}'.format(y_test.shape))
```

```python
#select a sample of the dataset
sample_df = scaling[['T2', 'T6']]

sample_df.head()
```

```python
#reshape sample dataset
x= sample_df['T2'].values.reshape(-1,1)
y = sample_df['T6'].values.reshape(-1,1)
```

```python
#split sample dataset into train and test sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#linear model on sample of dataset
from sklearn import linear_model
reg = linear_model.LinearRegression()

# Train the model using the training sets
reg.fit(xtrain, ytrain)

# Make predictions using the testing set
pred = reg.predict(xtest)
```

```python
#R-squared or Coefficient of determination
from sklearn.metrics import r2_score

r2_score = r2_score(ytest, pred)
print('R-squared:',(round(r2_score, 2)))
```

```python
#linear regression
from sklearn import linear_model

# Create linear regression object
linear = linear_model.LinearRegression()

# Train the model using the train set
linear.fit(X_train, y_train)

# predict using the test set
energy_pred = linear.predict(X_test)

print("Trainset score: {:.3f}".format(linear.score(X_train, y_train)))
print("Testset score: {:.3f}".format(linear.score(X_test, y_test)))
```

```python
#mean absolute erro
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, energy_pred)

print('MAE:',(round(mae, 2)))

#residual sum of squares
import numpy as np
rss = np.sum(np.square(y_test - energy_pred))
print('RSS:',(round(rss, 2)))

#root mean squared error
from sklearn.metrics import  mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, energy_pred))
print('RMSE:',(round(rmse, 3)))

#R-squared or coefficient of determination
from sklearn.metrics import r2_score

r2_score = r2_score(y_test, energy_pred)
print('R-squared:',(round(r2_score, 2)))
```

```python
#checking the effect of regularisation

def get_weights_df(model, feat, col_name):
    
  #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    
    return weights_df
```

```python
#weights of linear model
linear_model_weights = get_weights_df(linear, X_train, 'Linear_Model_Weight')
linear_model_weights
```

The RH_2 feature has the least weight while RH_1 feature has the highest weight

## Ridge Regression (L2)

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)

#obtain predictions
ridge_pred = ridge_reg.predict(X_test)

print("Trainset score: {:.3f}".format(ridge_reg.score(X_train, y_train)))
print("Testset score: {:.3f}".format(ridge_reg.score(X_test, y_test)))
```

```python
#root mean squared error
from sklearn.metrics import  mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))
print('RMSE:',(round(rmse, 3)))
```

There is no change in the root mean squared error since the rmse for the linear model is also 0.088

##Lasso Regression

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X_train, y_train)

#obtain predictions
lasso_pred = lasso_reg.predict(X_test)

print("Trainset score: {:.3f}".format(lasso_reg.score(X_train, y_train)))
print("Testset score: {:.3f}".format(lasso_reg.score(X_test, y_test)))
```

```python
#Lasso regression weights
lasso_weights_df = get_weights_df(lasso_reg, X_train, 'Lasso_weight')
lasso_weights_df
```

4features have non- zero weights. They are RH_out, RH_8, Windspeed and RH_1
